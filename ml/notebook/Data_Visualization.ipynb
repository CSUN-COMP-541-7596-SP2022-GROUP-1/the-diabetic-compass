```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import pie, axis, show

df = pd.read_csv ('diabetes_012_health_indicators_BRFSS2015.csv')

Diabetes_012 = df['Diabetes_012'].value_counts()
Diabetes_012.plot.pie(autopct='%1.0f%%')

title = plt.title('0 = no diabetes, 1 = prediabetes, 2 = diabetes')
plt.gca().axis("equal")

plt.show()
print(Diabetes_012)
```

```python
#https://stackoverflow.com/questions/33497559/display-a-histogram-with-very-non-uniform-bin-widths
#major bin allignment issue, resolved with the article above
import numpy as np
bins=[0,18.5,25,30,98]
data = df['BMI']

hist, bin_edges = np.histogram(data,bins)
fig,ax = plt.subplots()
ax.set_title('Body Mass Index')

ax.bar(range(len(hist)),hist,width=1)
ax.set_xticks([i for i,j in enumerate(hist)])
ax.set_xticklabels(['{} - {}'.format(bins[i],bins[i+1]) for i,j in enumerate(hist)])

plt.show()
print('<18.5 Underweight, <25 Normal,\n<30 Overweight, 30>= Obese')
```
